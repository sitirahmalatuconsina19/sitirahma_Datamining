```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DatasetCuaca_convert.csv')
```

```python
mean = df['Suhu_Udara'].mean()
median = df['Suhu_Udara'].median()
modus = df['Suhu_Udara'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus)
```

```python
mean = df['Curah_Hujan'].mean()
median = df['Curah_Hujan'].median()
modus = df['Curah_Hujan'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus) 
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Kelembapan_Udara'])
plt.title('Boxplot Kelembapan Udara')

plt.subplot(1, 2, 2)
plt.boxplot(df['Curah_Hujan']) 
plt.title('Boxplot Curah Hujan')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(data['Curah_Hujan'], data['Kelembapan_Udara'])
plt.xlabel('Curah Hujan')
plt.ylabel('Kelembapan Udara')
plt.title('Scatterplot Curah Hujan vs. Kelembapan Udara')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Curah_Hujan'], bins=10, edgecolor='black')
plt.xlabel('Curah Hujan')
plt.ylabel('frekuensi')
plt.title('Histogram Curah Hujan')

mean_curah_hujan = df['Curah_Hujan'].mean()
plt.axvline(mean_curah_hujan, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_curah_hujan+1, 20, f'Mean: {mean_curah_hujan:.2f}', color='red')

std_curah_hujan = df['Curah_Hujan'].std()
plt.axvline(mean_curah_hujan+std_curah_hujan, color='brown',linestyle='dashed',linewidth=1, label='Std Dev')
plt.axvline(mean_curah_hujan-std_curah_hujan, color='brown',linestyle='dashed',linewidth=1)
plt.text(mean_curah_hujan+std_curah_hujan+1, 15, f'Std Dev: {std_curah_hujan:.2f}', color='black')

plt.legend()
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(df['Suhu_Udara'], bins=10, edgecolor='black')
plt.xlabel('Suhu')
plt.ylabel('frekuensi')
plt.title('Histogram Suhu')

mean_suhu = df['Suhu_Udara'].mean()
plt.axvline(mean_suhu, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_suhu+1, 20, f'Mean: {mean_suhu:.2f}', color='red')

std_suhu = df['Suhu_Udara'].std()
plt.axvline(mean_suhu+std_suhu, color='brown', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_suhu-std_suhu, color='brown', linestyle='dashed', linewidth=1)
plt.text(mean_suhu+std_suhu+1, 15, f'Std Dev: {std_suhu:.2f}', color='brown')

plt.legend()
plt.show()
```
