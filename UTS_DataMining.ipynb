```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DatasetCuaca_convert.csv')
```

```python
mean_suhu = data['Suhu Udara'].mean()
median_suhu = data['Suhu Udara'].median()
modus_suhu = data['Suhu Udara'].mode()
```

```python
print("Mean Suhu Udara", mean_suhu)
print("Median Suhu Udara", median_suhu)
print("Modus Suhu Udara", modus_suhu)
```

```python
mean_curah_hujan = data['Curah Hujan'].mean()
median_curah_hujan = data['Curah Hujan'].median()
modus_curah_hujan = data['Curah Hujan'].mode()
```

```python
print("Mean Curah Hujan", mean_curah_hujan)
print("Median Curah Hujan", median_curah_hujan)
print("Modus Curah Hujan", modus_curah_hujan)
```

```python
plt.boxplot(data['Kelembapan Udara'])
plt.xlabel('Kelembapan Udara')
plt.ylabel('Nilai')
plt.title('Boxplot Kelembapan Udara')
plt.show()
```

```python
plt.scatter(data['Curah Hujan'], data['Kelembapan Udara'])
plt.xlabel('Curah Hujan')
plt.ylabel('Kelembapan Udara')
plt.title('Scatterplot Curah Hujan vs. Kelembapan Udara')
plt.show()
```

```python
plt.hist(data['Curah Hujan'], bins=10)
plt.xlabel('Curah Hujan')
plt.ylabel('Frekuensi')
plt.title('Histogram Curah Hujan')
plt.show()
```

```python
plt.hist(data['Suhu Udara'], bins=10)
plt.xlabel('Suhu Udara')
plt.ylabel('Frekuensi')
plt.title('Histogram Suhu Udara')
plt.show()
```
